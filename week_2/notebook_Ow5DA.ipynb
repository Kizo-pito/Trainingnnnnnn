```python
# Topic 1: Perceptron Learning Algorithm (PLA)
display(HTML("""
<div style="background: linear-gradient(135deg, #FF6B6B 0%, #FF8E53 100%); color: white; padding: 20px; border-radius: 15px; margin: 20px 0;">
    <h1 style="margin: 0; font-size: 2.5em;">🧠 Topic 1: Perceptron Learning Algorithm (PLA)</h1>
    <p style="font-size: 1.3em; margin: 10px 0 0 0; opacity: 0.9;">The Foundation of Linear Classification</p>
</div>
"""))

display(Markdown(r"""
## 📖 Mathematical Foundation

The **Perceptron Learning Algorithm (PLA)** is the simplest form of neural network and the foundation of linear classification. It was introduced by Frank Rosenblatt in 1957.

### 🔢 Mathematical Formulation

Given a dataset $D = \{(\mathbf{x}_1, y_1), (\mathbf{x}_2, y_2), \ldots, (\mathbf{x}_N, y_N)\}$ where:
- $\mathbf{x}_i \in \mathbb{R}^d$ is the input vector
- $y_i \in \{-1, +1\}$ is the binary label

The perceptron hypothesis is:
$$h(\mathbf{x}) = \text{sign}(\mathbf{w}^T\mathbf{x} + b) = \text{sign}(\sum_{i=1}^{d} w_i x_i + b)$$

Where:
- $\mathbf{w} = [w_1, w_2, \ldots, w_d]^T$ is the weight vector
- $b$ is the bias term
- $\text{sign}(z) = \begin{cases} +1 & \text{if } z > 0 \\ -1 & \text{if } z < 0 \end{cases}$

### 🎯 Learning Rule

The PLA updates weights when a misclassification occurs:

$$\mathbf{w}_{t+1} = \mathbf{w}_t + y_i \mathbf{x}_i$$
$$b_{t+1} = b_t + y_i$$

This update is performed only when $y_i \neq h(\mathbf{x}_i)$ (misclassification).

### 🏆 Convergence Theorem

**Rosenblatt's Convergence Theorem**: If the data is linearly separable, PLA will converge to a solution in finite steps.

The maximum number of updates is bounded by:
$$T \leq \frac{R^2}{\rho^2}$$

Where:
- $R = \max_i ||\mathbf{x}_i||$ (maximum norm of input vectors)
- $\rho$ is the margin of separation
"""))
```

```python
# Complete implementation of Perceptron Learning Algorithm
class Perceptron:
    """
    Complete implementation of the Perceptron Learning Algorithm (PLA)
    
    Features:
    - Mathematical foundation implementation
    - Convergence tracking
    - Visualization capabilities
    - Multiple initialization strategies
    """
    
    def __init__(self, learning_rate=1.0, max_iter=1000, random_state=42):
        """
        Initialize Perceptron
        
        Parameters:
        - learning_rate: Step size for weight updates
        - max_iter: Maximum number of iterations
        - random_state: Random seed for reproducibility
        """
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.random_state = random_state
        self.weights = None
        self.bias = None
        self.errors_history = []
        self.weights_history = []
        self.convergence_step = None
        
    def _add_bias_column(self, X):
        """Add bias column to input matrix"""
        return np.c_[np.ones(X.shape[0]), X]
    
    def _sign(self, z):
        """Sign activation function"""
        return np.where(z >= 0, 1, -1)
    
    def fit(self, X, y, verbose=True):
        """
        Train the perceptron using PLA
        
        Parameters:
        - X: Training features (n_samples, n_features)
        - y: Training labels {-1, +1}
        - verbose: Print training progress
        """
        # Initialize random generator
        np.random.seed(self.random_state)
        
        # Add bias term to features
        X_with_bias = self._add_bias_column(X)
        n_samples, n_features = X_with_bias.shape
        
        # Initialize weights (including bias)
        self.weights = np.random.normal(0, 0.01, n_features)
        
        # Track training history
        self.errors_history = []
        self.weights_history = [self.weights.copy()]
        
        if verbose:
            print("🚀 Starting Perceptron Learning Algorithm...")
            print(f"📊 Dataset: {n_samples} samples, {n_features-1} features")
            print(f"⚙️  Parameters: lr={self.learning_rate}, max_iter={self.max_iter}")
            print("-" * 60)
        
        # Training loop
        for iteration in range(self.max_iter):
            errors = 0
            
            # Iterate through all training examples
            for i in range(n_samples):
                # Compute prediction
                prediction = self._sign(np.dot(X_with_bias[i], self.weights))
                
                # Check for misclassification
                if prediction != y[i]:
                    # PLA update rule: w = w + y_i * x_i
                    self.weights += self.learning_rate * y[i] * X_with_bias[i]
                    errors += 1
            
            # Record training progress
            self.errors_history.append(errors)
            self.weights_history.append(self.weights.copy())
            
            if verbose and (iteration + 1) % 50 == 0:
                print(f"Iteration {iteration + 1:4d}: {errors:3d} misclassifications")
            
            # Check for convergence
            if errors == 0:
                self.convergence_step = iteration + 1
                if verbose:
                    print(f"✅ Converged after {self.convergence_step} iterations!")
                break
        else:
            if verbose:
                print(f"⚠️  Did not converge after {self.max_iter} iterations")
                print(f"   Final misclassifications: {self.errors_history[-1]}")
        
        return self
    
    def predict(self, X):
        """Make predictions on new data"""
        X_with_bias = self._add_bias_column(X)
        return self._sign(np.dot(X_with_bias, self.weights))
    
    def decision_function(self, X):
        """Compute decision function values"""
        X_with_bias = self._add_bias_column(X)
        return np.dot(X_with_bias, self.weights)
    
    def score(self, X, y):
        """Compute accuracy"""
        predictions = self.predict(X)
        return np.mean(predictions == y)

# Test the implementation
print("✅ Perceptron class implemented successfully!")
print("📊 Ready for training and visualization...")
```

```python
# Create linearly separable dataset for PLA demonstration
def create_linearly_separable_data(n_samples=100, random_state=42):
    """Create a linearly separable 2D dataset"""
    np.random.seed(random_state)
    
    # Generate two clusters
    cluster1 = np.random.multivariate_normal([2, 2], [[0.5, 0.1], [0.1, 0.5]], n_samples//2)
    cluster2 = np.random.multivariate_normal([-1, -1], [[0.5, -0.1], [-0.1, 0.5]], n_samples//2)
    
    # Combine data
    X = np.vstack([cluster1, cluster2])
    y = np.hstack([np.ones(n_samples//2), -np.ones(n_samples//2)])
    
    # Shuffle the data
    indices = np.random.permutation(n_samples)
    X, y = X[indices], y[indices]
    
    return X, y

# Generate training data
X_train, y_train = create_linearly_separable_data(n_samples=100, random_state=42)

print("📊 Generated linearly separable dataset:")
print(f"   Features shape: {X_train.shape}")
print(f"   Labels shape: {y_train.shape}")
print(f"   Positive examples: {np.sum(y_train == 1)}")
print(f"   Negative examples: {np.sum(y_train == -1)}")

# Visualize the dataset
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
mask_pos = y_train == 1
mask_neg = y_train == -1
plt.scatter(X_train[mask_pos, 0], X_train[mask_pos, 1], 
           c='red', marker='o', s=50, alpha=0.7, label='Class +1')
plt.scatter(X_train[mask_neg, 0], X_train[mask_neg, 1], 
           c='blue', marker='s', s=50, alpha=0.7, label='Class -1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('🎯 Training Dataset\n(Linearly Separable)')
plt.legend()
plt.grid(True, alpha=0.3)

# Train the perceptron
perceptron = Perceptron(learning_rate=1.0, max_iter=100, random_state=42)
perceptron.fit(X_train, y_train, verbose=True)

# Plot decision boundary
plt.subplot(1, 2, 2)
plt.scatter(X_train[mask_pos, 0], X_train[mask_pos, 1], 
           c='red', marker='o', s=50, alpha=0.7, label='Class +1')
plt.scatter(X_train[mask_neg, 0], X_train[mask_neg, 1], 
           c='blue', marker's', s=50, alpha=0.7, label='Class -1')

# Plot decision boundary
if perceptron.weights is not None:
    # Decision boundary: w0 + w1*x1 + w2*x2 = 0
    # Solve for x2: x2 = -(w0 + w1*x1) / w2
    w0, w1, w2 = perceptron.weights
    x1_range = np.linspace(X_train[:, 0].min()-1, X_train[:, 0].max()+1, 100)
    x2_boundary = -(w0 + w1 * x1_range) / w2
    plt.plot(x1_range, x2_boundary, 'g-', linewidth=2, label='Decision Boundary')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('🎯 PLA Result\n(Decision Boundary)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print results
accuracy = perceptron.score(X_train, y_train)
print(f"\n📈 Training Results:")
print(f"   Final accuracy: {accuracy:.2%}")
print(f"   Convergence: {'Yes' if perceptron.convergence_step else 'No'}")
if perceptron.convergence_step:
    print(f"   Converged in: {perceptron.convergence_step} iterations")
print(f"   Final weights: {perceptron.weights}")
```

```python
# Topic 2: Logistic Regression
display(HTML("""
<div style="background: linear-gradient(135deg, #4ECDC4 0%, #44A08D 100%); color: white; padding: 20px; border-radius: 15px; margin: 20px 0;">
    <h1 style="margin: 0; font-size: 2.5em;">📈 Topic 2: Logistic Regression</h1>
    <p style="font-size: 1.3em; margin: 10px 0 0 0; opacity: 0.9;">Probabilistic Classification with Gradient Descent</p>
</div>
"""))

display(Markdown(r"""
## 📖 Mathematical Foundation

**Logistic Regression** extends linear regression to classification by using the logistic (sigmoid) function to model probabilities. Unlike PLA, it provides probabilistic outputs and uses gradient descent for optimization.

### 🔢 Mathematical Formulation

Given a dataset $D = \{(\mathbf{x}_1, y_1), (\mathbf{x}_2, y_2), \ldots, (\mathbf{x}_N, y_N)\}$ where:
- $\mathbf{x}_i \in \mathbb{R}^d$ is the input vector
- $y_i \in \{0, 1\}$ is the binary label (note: using 0,1 instead of -1,+1)

### 🎯 Sigmoid Function

The logistic function (sigmoid) maps real numbers to probabilities:
$$\sigma(z) = \frac{1}{1 + e^{-z}} = \frac{e^z}{1 + e^z}$$

Properties of the sigmoid function:
- $\sigma(z) \in (0, 1)$ for all $z \in \mathbb{R}$
- $\sigma(0) = 0.5$
- $\sigma(-z) = 1 - \sigma(z)$
- $\frac{d\sigma}{dz} = \sigma(z)(1 - \sigma(z))$

### 📊 Logistic Regression Model

The hypothesis function is:
$$h_{\mathbf{w}}(\mathbf{x}) = \sigma(\mathbf{w}^T\mathbf{x} + b) = \frac{1}{1 + e^{-(\mathbf{w}^T\mathbf{x} + b)}}$$

This represents:
$$P(y = 1 | \mathbf{x}) = h_{\mathbf{w}}(\mathbf{x})$$
$$P(y = 0 | \mathbf{x}) = 1 - h_{\mathbf{w}}(\mathbf{x})$$

### 📉 Cost Function (Log-Likelihood)

The cost function is derived from maximum likelihood estimation:
$$J(\mathbf{w}) = -\frac{1}{N} \sum_{i=1}^{N} \left[ y_i \log(h_{\mathbf{w}}(\mathbf{x}_i)) + (1-y_i) \log(1-h_{\mathbf{w}}(\mathbf{x}_i)) \right]$$

This is the **cross-entropy loss** or **log-loss**.

### 🎯 Gradient Computation

The gradient of the cost function with respect to weights:
$$\frac{\partial J}{\partial \mathbf{w}} = \frac{1}{N} \sum_{i=1}^{N} (h_{\mathbf{w}}(\mathbf{x}_i) - y_i) \mathbf{x}_i$$

$$\frac{\partial J}{\partial b} = \frac{1}{N} \sum_{i=1}^{N} (h_{\mathbf{w}}(\mathbf{x}_i) - y_i)$$

### 🔄 Update Rule (Gradient Descent)

$$\mathbf{w}_{t+1} = \mathbf{w}_t - \alpha \frac{\partial J}{\partial \mathbf{w}}$$
$$b_{t+1} = b_t - \alpha \frac{\partial J}{\partial b}$$

Where $\alpha$ is the learning rate.
"""))
```

```python
# Complete implementation of Logistic Regression
class LogisticRegression:
    """
    Complete implementation of Logistic Regression with Gradient Descent
    
    Features:
    - Mathematical foundation implementation
    - Cost function tracking
    - Probabilistic predictions
    - Visualization capabilities
    """
    
    def __init__(self, learning_rate=0.01, max_iter=1000, tolerance=1e-6, random_state=42):
        """
        Initialize Logistic Regression
        
        Parameters:
        - learning_rate: Step size for gradient descent
        - max_iter: Maximum number of iterations
        - tolerance: Convergence tolerance
        - random_state: Random seed for reproducibility
        """
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.tolerance = tolerance
        self.random_state = random_state
        self.weights = None
        self.cost_history = []
        self.weights_history = []
        
    def _add_bias_column(self, X):
        """Add bias column to input matrix"""
        return np.c_[np.ones(X.shape[0]), X]
    
    def _sigmoid(self, z):
        """Sigmoid activation function with numerical stability"""
        # Clip z to prevent overflow
        z = np.clip(z, -250, 250)
        return 1 / (1 + np.exp(-z))
    
    def _cost_function(self, h, y):
        """Compute cross-entropy cost function"""
        # Add small epsilon to prevent log(0)
        epsilon = 1e-15
        h = np.clip(h, epsilon, 1 - epsilon)
        return -np.mean(y * np.log(h) + (1 - y) * np.log(1 - h))
    
    def fit(self, X, y, verbose=True):
        """
        Train the logistic regression using gradient descent
        
        Parameters:
        - X: Training features (n_samples, n_features)
        - y: Training labels {0, 1}
        - verbose: Print training progress
        """
        # Initialize random generator
        np.random.seed(self.random_state)
        
        # Add bias term to features
        X_with_bias = self._add_bias_column(X)
        n_samples, n_features = X_with_bias.shape
        
        # Initialize weights
        self.weights = np.random.normal(0, 0.01, n_features)
        
        # Track training history
        self.cost_history = []
        self.weights_history = [self.weights.copy()]
        
        if verbose:
            print("🚀 Starting Logistic Regression Training...")
            print(f"📊 Dataset: {n_samples} samples, {n_features-1} features")
            print(f"⚙️  Parameters: lr={self.learning_rate}, max_iter={self.max_iter}")
            print("-" * 60)
        
        # Training loop
        for iteration in range(self.max_iter):
            # Forward pass: compute predictions
            z = np.dot(X_with_bias, self.weights)
            h = self._sigmoid(z)
            
            # Compute cost
            cost = self._cost_function(h, y)
            self.cost_history.append(cost)
            
            # Compute gradients
            dw = np.dot(X_with_bias.T, (h - y)) / n_samples
            
            # Update weights
            new_weights = self.weights - self.learning_rate * dw
            
            # Check for convergence
            if np.linalg.norm(new_weights - self.weights) < self.tolerance:
                if verbose:
                    print(f"✅ Converged after {iteration + 1} iterations!")
                break
            
            self.weights = new_weights
            self.weights_history.append(self.weights.copy())
            
            if verbose and (iteration + 1) % 100 == 0:
                print(f"Iteration {iteration + 1:4d}: Cost = {cost:.6f}")
        
        else:
            if verbose:
                print(f"⚠️  Reached maximum iterations ({self.max_iter})")
                print(f"   Final cost: {self.cost_history[-1]:.6f}")
        
        return self
    
    def predict_proba(self, X):
        """Predict class probabilities"""
        X_with_bias = self._add_bias_column(X)
        return self._sigmoid(np.dot(X_with_bias, self.weights))
    
    def predict(self, X, threshold=0.5):
        """Make binary predictions"""
        return (self.predict_proba(X) >= threshold).astype(int)
    
    def score(self, X, y):
        """Compute accuracy"""
        predictions = self.predict(X)
        return np.mean(predictions == y)

# Demonstrate sigmoid function properties
print("🔍 Visualizing Sigmoid Function Properties...")

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Plot 1: Sigmoid function
z = np.linspace(-10, 10, 100)
sigmoid_z = 1 / (1 + np.exp(-z))

axes[0].plot(z, sigmoid_z, 'b-', linewidth=3, label='σ(z)')
axes[0].axhline(y=0.5, color='r', linestyle='--', alpha=0.7, label='Decision boundary')
axes[0].axvline(x=0, color='r', linestyle='--', alpha=0.7)
axes[0].set_xlabel('z')
axes[0].set_ylabel('σ(z)')
axes[0].set_title('🎯 Sigmoid Function\nσ(z) = 1/(1+e⁻ᶻ)')
axes[0].grid(True, alpha=0.3)
axes[0].legend()

# Plot 2: Sigmoid derivative
sigmoid_derivative = sigmoid_z * (1 - sigmoid_z)
axes[1].plot(z, sigmoid_derivative, 'g-', linewidth=3, label="σ'(z)")
axes[1].set_xlabel('z')
axes[1].set_ylabel("σ'(z)")
axes[1].set_title("📊 Sigmoid Derivative\nσ'(z) = σ(z)(1-σ(z))")
axes[1].grid(True, alpha=0.3)
axes[1].legend()

# Plot 3: Comparison with other activation functions
tanh_z = np.tanh(z)
relu_z = np.maximum(0, z)

axes[2].plot(z, sigmoid_z, 'b-', linewidth=2, label='Sigmoid')
axes[2].plot(z, tanh_z, 'r-', linewidth=2, label='Tanh')
axes[2].plot(z, relu_z, 'g-', linewidth=2, label='ReLU')
axes[2].set_xlabel('z')
axes[2].set_ylabel('Activation')
axes[2].set_title('🔄 Activation Functions\nComparison')
axes[2].grid(True, alpha=0.3)
axes[2].legend()

plt.tight_layout()
plt.show()

print("✅ Logistic Regression class implemented successfully!")
print("📊 Ready for training and comparison with PLA...")
```

```python
# Topic 3: Overfitting
display(HTML("""
<div style="background: linear-gradient(135deg, #FF9A9E 0%, #FECFEF 50%, #FECFEF 100%); color: #333; padding: 20px; border-radius: 15px; margin: 20px 0;">
    <h1 style="margin: 0; font-size: 2.5em; color: #2c3e50;">⚠️ Topic 3: Overfitting</h1>
    <p style="font-size: 1.3em; margin: 10px 0 0 0; opacity: 0.8; color: #34495e;">Model Complexity, Generalization & Prevention Techniques</p>
</div>
"""))

display(Markdown(r"""
## 📖 Understanding Overfitting

**Overfitting** occurs when a model learns the training data too well, including noise and random fluctuations, resulting in poor generalization to new, unseen data.

### 🎯 Key Concepts

#### 1. **Bias-Variance Tradeoff**
The total error can be decomposed as:
$$\text{Total Error} = \text{Bias}^2 + \text{Variance} + \text{Irreducible Error}$$

Where:
- **Bias**: Error due to overly simplistic assumptions
- **Variance**: Error due to sensitivity to small fluctuations in training set
- **Irreducible Error**: Inherent noise in the data

#### 2. **Model Complexity**
As model complexity increases:
- **Training Error**: Generally decreases
- **Validation Error**: Initially decreases, then increases (U-shaped curve)
- **Optimal Complexity**: Minimum validation error point

#### 3. **Learning Curves**
- **Underfitting**: High bias, low variance (both training and validation errors are high)
- **Good Fit**: Balanced bias and variance (training and validation errors are close and low)
- **Overfitting**: Low bias, high variance (large gap between training and validation errors)

### 🛡️ Prevention Techniques

1. **Cross-Validation**: Use k-fold CV to estimate generalization performance
2. **Regularization**: Add penalty terms to prevent complex models
   - **L1 Regularization (Lasso)**: $\lambda \sum_{i=1}^{n} |w_i|$
   - **L2 Regularization (Ridge)**: $\lambda \sum_{i=1}^{n} w_i^2$
3. **Early Stopping**: Stop training when validation error starts increasing
4. **Dropout**: Randomly set neurons to zero during training (for neural networks)
5. **Data Augmentation**: Increase training data diversity
6. **Ensemble Methods**: Combine multiple models to reduce variance
"""))
```

```python
# Demonstrate overfitting with polynomial regression
def create_noisy_dataset(n_samples=100, noise_level=0.3, random_state=42):
    """Create a dataset with underlying pattern plus noise"""
    np.random.seed(random_state)
    
    # Generate true function: f(x) = 0.5x^3 - 2x^2 + x + 1
    X = np.linspace(-2, 2, n_samples)
    y_true = 0.5 * X**3 - 2 * X**2 + X + 1
    
    # Add noise
    noise = np.random.normal(0, noise_level, n_samples)
    y_noisy = y_true + noise
    
    return X.reshape(-1, 1), y_noisy, y_true

def create_polynomial_features(X, degree):
    """Create polynomial features up to given degree"""
    n_samples = X.shape[0]
    features = np.ones((n_samples, 1))  # bias term
    
    for d in range(1, degree + 1):
        features = np.c_[features, X**d]
    
    return features

def fit_polynomial(X, y, degree):
    """Fit polynomial regression using normal equation"""
    # Create polynomial features
    X_poly = create_polynomial_features(X, degree)
    
    # Solve using normal equation: w = (X^T X)^(-1) X^T y
    try:
        weights = np.linalg.solve(X_poly.T @ X_poly, X_poly.T @ y)
    except np.linalg.LinAlgError:
        # Use pseudo-inverse if matrix is singular
        weights = np.linalg.pinv(X_poly.T @ X_poly) @ X_poly.T @ y
    
    return weights

def predict_polynomial(X, weights):
    """Make predictions using polynomial weights"""
    degree = len(weights) - 1
    X_poly = create_polynomial_features(X, degree)
    return X_poly @ weights

# Generate dataset
X_train, y_train, y_true = create_noisy_dataset(n_samples=50, noise_level=0.5, random_state=42)
X_test, y_test, y_true_test = create_noisy_dataset(n_samples=30, noise_level=0.5, random_state=123)

# Test different polynomial degrees
degrees = [1, 2, 3, 5, 8, 15]
train_errors = []
test_errors = []
models = []

print("🔍 Testing different polynomial degrees for overfitting demonstration...")
print("-" * 70)

for degree in degrees:
    # Fit model
    weights = fit_polynomial(X_train, y_train, degree)
    models.append(weights)
    
    # Calculate errors
    train_pred = predict_polynomial(X_train, weights)
    test_pred = predict_polynomial(X_test, weights)
    
    train_mse = np.mean((y_train - train_pred)**2)
    test_mse = np.mean((y_test - test_pred)**2)
    
    train_errors.append(train_mse)
    test_errors.append(test_mse)
    
    print(f"Degree {degree:2d}: Train MSE = {train_mse:.4f}, Test MSE = {test_mse:.4f}")

print("✅ Model fitting complete!")

# Create comprehensive visualization
plt.figure(figsize=(20, 12))

# Plot 1: Training data and true function
plt.subplot(2, 4, 1)
X_plot = np.linspace(-2, 2, 200).reshape(-1, 1)
y_true_plot = 0.5 * X_plot.ravel()**3 - 2 * X_plot.ravel()**2 + X_plot.ravel() + 1

plt.scatter(X_train, y_train, color='blue', alpha=0.6, s=50, label='Training Data')
plt.plot(X_plot, y_true_plot, 'r-', linewidth=3, label='True Function')
plt.xlabel('X')
plt.ylabel('y')
plt.title('📊 Training Data\nvs True Function')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 2-4: Different model complexities
for i, degree in enumerate([1, 3, 15]):
    plt.subplot(2, 4, i + 2)
    
    # Plot data
    plt.scatter(X_train, y_train, color='blue', alpha=0.6, s=50, label='Training Data')
    plt.plot(X_plot, y_true_plot, 'r-', linewidth=2, alpha=0.7, label='True Function')
    
    # Plot model prediction
    model_pred = predict_polynomial(X_plot, models[degrees.index(degree)])
    plt.plot(X_plot, model_pred, 'g-', linewidth=3, label=f'Degree {degree} Model')
    
    plt.xlabel('X')
    plt.ylabel('y')
    plt.title(f'🎯 Polynomial Degree {degree}\n{"Underfitting" if degree == 1 else "Good Fit" if degree == 3 else "Overfitting"}')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.ylim(-5, 5)

# Plot 5: Learning curves (Error vs Model Complexity)
plt.subplot(2, 4, 5)
plt.plot(degrees, train_errors, 'bo-', linewidth=2, markersize=8, label='Training Error')
plt.plot(degrees, test_errors, 'ro-', linewidth=2, markersize=8, label='Test Error')
plt.xlabel('Polynomial Degree (Model Complexity)')
plt.ylabel('Mean Squared Error')
plt.title('📈 Model Complexity\nvs Error (Overfitting Curve)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.yscale('log')

# Find optimal complexity
optimal_idx = np.argmin(test_errors)
optimal_degree = degrees[optimal_idx]
plt.axvline(x=optimal_degree, color='green', linestyle='--', alpha=0.7, 
           label=f'Optimal (Degree {optimal_degree})')
plt.legend()

# Plot 6: Bias-Variance Illustration
plt.subplot(2, 4, 6)
# Simulate bias-variance tradeoff
complexity_range = np.linspace(1, 15, 50)
bias_squared = 2 * np.exp(-complexity_range/3) + 0.1  # Decreases with complexity
variance = 0.1 * complexity_range**1.5 / 15  # Increases with complexity
total_error = bias_squared + variance + 0.5  # Add irreducible error

plt.plot(complexity_range, bias_squared, 'r-', linewidth=2, label='Bias²')
plt.plot(complexity_range, variance, 'b-', linewidth=2, label='Variance')
plt.plot(complexity_range, total_error, 'g-', linewidth=3, label='Total Error')
plt.axhline(y=0.5, color='black', linestyle=':', alpha=0.7, label='Irreducible Error')

plt.xlabel('Model Complexity')
plt.ylabel('Error')
plt.title('🎯 Bias-Variance Tradeoff\n(Theoretical)')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 7: Weight magnitudes vs complexity
plt.subplot(2, 4, 7)
weight_norms = []
for degree, weights in zip(degrees, models):
    weight_norm = np.linalg.norm(weights[1:])  # Exclude bias
    weight_norms.append(weight_norm)

plt.plot(degrees, weight_norms, 'mo-', linewidth=2, markersize=8)
plt.xlabel('Polynomial Degree')
plt.ylabel('||weights|| (L2 norm)')
plt.title('📊 Weight Magnitude\nvs Model Complexity')
plt.grid(True, alpha=0.3)
plt.yscale('log')

# Plot 8: Cross-validation demonstration
plt.subplot(2, 4, 8)
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

cv_scores_mean = []
cv_scores_std = []

for degree in degrees:
    # Create pipeline
    poly_model = Pipeline([
        ('poly', PolynomialFeatures(degree=degree)),
        ('linear', LinearRegression())
    ])
    
    # Perform cross-validation
    cv_scores = cross_val_score(poly_model, X_train, y_train, cv=5, 
                               scoring='neg_mean_squared_error')
    cv_scores = -cv_scores  # Convert to positive MSE
    
    cv_scores_mean.append(np.mean(cv_scores))
    cv_scores_std.append(np.std(cv_scores))

plt.errorbar(degrees, cv_scores_mean, yerr=cv_scores_std, 
             marker='o', linewidth=2, markersize=8, capsize=5)
plt.xlabel('Polynomial Degree')
plt.ylabel('Cross-Validation MSE')
plt.title('🔄 Cross-Validation\nScore vs Complexity')
plt.grid(True, alpha=0.3)
plt.yscale('log')

plt.tight_layout()
plt.show()

print("\n" + "="*80)
print("📊 OVERFITTING ANALYSIS RESULTS")
print("="*80)
print(f"🎯 Optimal Model Complexity: Degree {optimal_degree}")
print(f"📈 Training Error at Optimal: {train_errors[optimal_idx]:.4f}")
print(f"📉 Test Error at Optimal: {test_errors[optimal_idx]:.4f}")
print(f"\n🔍 Key Observations:")
print(f"   • Degree 1: Underfitting (high bias, low variance)")
print(f"   • Degree {optimal_degree}: Good balance (optimal tradeoff)")  
print(f"   • Degree 15: Overfitting (low bias, high variance)")
print(f"   • Weight magnitudes grow exponentially with complexity")
print(f"   • Cross-validation helps identify optimal complexity")
```

```python
# Topic 4: Gradient Descent (GD)
display(HTML("""
<div style="background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); color: white; padding: 20px; border-radius: 15px; margin: 20px 0;">
    <h1 style="margin: 0; font-size: 2.5em;">⬇️ Topic 4: Gradient Descent (GD)</h1>
    <p style="font-size: 1.3em; margin: 10px 0 0 0; opacity: 0.9;">The Foundation of Optimization in Machine Learning</p>
</div>
"""))

display(Markdown(r"""
## 📖 Mathematical Foundation

**Gradient Descent** is the cornerstone optimization algorithm in machine learning. It finds the minimum of a function by iteratively moving in the direction of steepest decrease.

### 🎯 Core Concept

For a function $f(\mathbf{w})$ where $\mathbf{w} \in \mathbb{R}^d$:

$$\mathbf{w}_{t+1} = \mathbf{w}_t - \alpha \nabla f(\mathbf{w}_t)$$

Where:
- $\mathbf{w}_t$ is the parameter vector at iteration $t$
- $\alpha$ is the learning rate (step size)
- $\nabla f(\mathbf{w}_t)$ is the gradient vector at $\mathbf{w}_t$

### 📐 Mathematical Intuition

The gradient $\nabla f(\mathbf{w})$ points in the direction of **steepest ascent**. By moving in the **opposite direction** ($-\nabla f(\mathbf{w})$), we move toward the minimum.

For a scalar function $f(w)$:
$$\nabla f(w) = \frac{df}{dw}$$

For a vector function $f(\mathbf{w}) = f(w_1, w_2, \ldots, w_d)$:
$$\nabla f(\mathbf{w}) = \begin{bmatrix} \frac{\partial f}{\partial w_1} \\ \frac{\partial f}{\partial w_2} \\ \vdots \\ \frac{\partial f}{\partial w_d} \end{bmatrix}$$

### 🔄 Algorithm Steps

1. **Initialize** parameters $\mathbf{w}_0$ (randomly or zeros)
2. **Repeat** until convergence:
   - Compute gradient: $\mathbf{g}_t = \nabla f(\mathbf{w}_t)$
   - Update parameters: $\mathbf{w}_{t+1} = \mathbf{w}_t - \alpha \mathbf{g}_t$
   - Check convergence: $||\mathbf{g}_t|| < \epsilon$ or $|f(\mathbf{w}_{t+1}) - f(\mathbf{w}_t)| < \epsilon$

### ⚙️ Learning Rate Selection

The learning rate $\alpha$ is crucial:
- **Too small**: Slow convergence
- **Too large**: May overshoot minimum or diverge
- **Adaptive methods**: Adjust $\alpha$ during training

### 🎭 Variants

1. **Batch GD**: Uses entire dataset for each update
2. **Mini-batch GD**: Uses subset of data
3. **Stochastic GD**: Uses single example
4. **Momentum**: Accelerates convergence
5. **Adam, RMSprop**: Adaptive learning rates
"""))
```
